OPTIMIZATION ATTEMPT 1

Preprocessing - Further Changes to Columns and Bins (only)

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns 'EIN' and 'NAME' plus the 'ASK_AMT' column.
application_df = application_df.drop(['EIN', 'NAME', 'ASK_AMT'], axis=1)
application_df.head()
```

Further analyze the columns

```python
application_df.nunique()
```

Reconsider APPLICATION_TYPE value counts for binning

```python
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Reduce APPLICATION_TYPE bins to 2
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts().index[1:17].tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

Reconsider CLASSIFICATION value counts for binning

```python
application_df['AFFILIATION'].value_counts()
```

```python
# Place AFFILIATION types into 3 bins
affiliations_to_replace = application_df['AFFILIATION'].value_counts().index[2:6].tolist()

# Replace in dataframe
for aff in affiliations_to_replace:
    application_df['AFFILIATION'] = application_df['AFFILIATION'].replace(aff,"Other")

# Check to make sure binning was successful
application_df['AFFILIATION'].value_counts()
```

```python
application_df['CLASSIFICATION'].value_counts()
```

```python
# Maintain current CLASSIFICATION value counts for binning
classifications_to_replace = application_df['CLASSIFICATION'].value_counts().index[5:71].tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df['USE_CASE'].value_counts()
```

```python
# Place USE_CASE types into 3 bins
use_cases_to_replace = application_df['USE_CASE'].value_counts().index[2:5].tolist()

# Replace in dataframe
for use in use_cases_to_replace:
    application_df['USE_CASE'] = application_df['USE_CASE'].replace(use,"Other")

# Check to make sure binning was successful
application_df['USE_CASE'].value_counts()
```

```python
application_df['ORGANIZATION'].value_counts()
```

```python
# Place ORGANIZATION types into 3 bins
orgs_to_replace = application_df['ORGANIZATION'].value_counts().index[2:5].tolist()

# Replace in dataframe
for orgs in orgs_to_replace:
    application_df['ORGANIZATION'] = application_df['ORGANIZATION'].replace(orgs,"Other")

# Check to make sure binning was successful
application_df['ORGANIZATION'].value_counts()
```

```python
application_df['INCOME_AMT'].value_counts()
```

```python
# Place Income Amounts into 4 bins
inc_to_replace = application_df['INCOME_AMT'].value_counts().index[3:9].tolist()

# Replace in dataframe
for inc in inc_to_replace:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(inc,"Other")

# Check to make sure binning was successful
application_df['INCOME_AMT'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'],
                          dtype=int)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns=['IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=1,
                                                    stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Hold variables to original model (except for input dimensions)

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(input_dim = 24, units = 80, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units = 30, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
